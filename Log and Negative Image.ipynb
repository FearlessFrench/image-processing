```python
import matplotlib.pyplot as plt
from PIL import Image

pdf_src_path = '/content/drive/MyDrive/Image Processing/Images/'
image = Image.open(pdf_src_path + 'IMG_7659.jpg')
```

### Log

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img_rgb = image.convert('RGB')
img_rgb.save(pdf_src_path + 'IMG_7659.png')
plt.imshow(img_rgb)
plt.show()

c = 255 / np.log(1+img_rgb.max())
c

img_tmp = 1. + img_rgb
print(img_tmp.min(), img_tmp.max())

img_log = c * np.log(1.+img_rgb)
print(img_log.min(), img_log.max())
img_log = np.array(img_log, dtype=np.uint8)
plt.imshow(img_log)
plt.show()
img_log_bgr = cv2.cvtColor(img_log, cv2.COLOR_RGB2BGR)
cv2.imwrite('IMG_7659.jpg', img_log_bgr)
```

```python

import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('Sky.jpg') # BGR
# cv2.imshow('Loaded image', img)
# cv2.waitKey(0)
# cv2.destroyAllWindows()

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # RGB
# img_arr = np.asarray(img_rgb)
plt.imshow(img_rgb)
plt.show()

print(img_rgb.min(), img_rgb.max())

img_neg = float(img_rgb.max()) - img_rgb
print(img_neg.min(), img_neg.max())
img_neg = np.array(img_neg, dtype=np.uint8)
plt.imshow(img_neg)
plt.show()
img_neg_bgr = cv2.cvtColor(img_neg, cv2.COLOR_RGB2BGR)
cv2.imwrite('Sky_Neg.jpg', img_neg_bgr)
```
